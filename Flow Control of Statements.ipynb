```python
"""
Write a program that gets three numbers from the user and displays the
biggest number among the three.
"""

def largest_number():
    while True:
        try:
            num1 = int(input('Enter your first number: '))
            num2 = int(input('Enter your second number: '))
            num3 = int(input('Enter your third number: '))
            break
        except ValueError:
            print('Invalid input. Please enter integer numbers.')
            continue

    numbers = [num1, num2, num3]
    largest_num = max(numbers)
    print(f"The largest number of your entries is {largest_num}.")

largest_number()
```

```python
"""
Compound interest is a common practice in finance and banking,
allowing you to earn interest on interest as well as on the principal.
Assume that your bank offers a savings account with which you can
earn compound interest. The amount you deposit into the account is
p, and annual compound interest is r. By the end of n years after your
initial deposit, the account balance will be a = p(1 + r)^n. For this project,
write a program that takes three numbers from the user as the initial
deposit, the annual interest, and the number of years that the user
wants the money to stay in the account. Calculate and display how
much money the user will have by the end of the nth year.
"""

def compound_growth():
    while True:
        try:
            # Input the principal amount
            principal = float(input('Please enter the principal of your investment: '))
            if principal <= 0:
                print('Principal must be a positive number. Please try again.')
                continue
            
            # Input the annual interest rate
            interest = float(input('Please enter the annual interest rate as a decimal (e.g., 0.05 for 5%): '))
            if interest < 0 or interest >= 1:
                print('Please enter an interest rate between 0 and 1 (e.g., 0.05 for 5%).')
                continue
            
            # Input the number of years
            years = int(input('Please enter the length of your investment in years: '))
            if years <= 0:
                print('Number of years must be a positive integer. Please try again.')
                continue

            # Calculate the total savings
            total_savings = principal * (1 + interest) ** years

            # Display the result
            print(f'The amount of total savings from your investment is ${total_savings:.2f}, '
                  f'given your principal of ${principal:.2f}, an interest rate of {interest*100:.2f}%, '
                  f'and a duration of {years} years.')
            
            break

        except ValueError:
            print('Invalid input! Please enter numerical values for principal, interest rate, and number of years.')

compound_growth()
```

```python
"""
A mortgage is the money borrowed from a lender for the purchase
of a property. Mortgages and mortgage payments are a big thing for
almost everyone. For a mortgage with principal of P at a fixed monthly
interest rate of r that needs to be paid off in Y years or 12 * Y = N
months, the monthly payment would be:
Write a program that takes the principal, fixed annual interest rate,
and years of amortization then calculates and displays the monthly
payment amount. Hint: You will need to work out the number of
months from number of years, and the monthly interest rate from the
annual interest rate.
"""

def monthly_mortgage_payment():
    flag = True
    while flag == True:
        try:
            p = float(input('Enter your mortgage principal without commas or dollar signs: '))
            if p <= 0:
                print("Principal cannot be less or equal to 0.")
                continue
            ir = float(input('Enter your interest rate (0 <= rate <= 1): '))
            if ir < 0 or ir > 1:
                print("Your interest rate should be between 0 and 1.")
                continue

            amor = float(input('Enter your amortization in years: '))
            if amor <= 0:
                print("Your amortization period cannot be less or equal to 0.")
                continue
            flag = False
            break

        except ValueError:
            print('Wrong input.')
            continue

    monthly_ir = ir/12
    amor_in_months = amor * 12
    nominator = (monthly_ir * p) * (1 + monthly_ir) ** amor_in_months
    denominator = ((1 + monthly_ir) ** amor_in_months) - 1
    payment = nominator / denominator
    print(f'Your approximate monthly mortgage payment is ${round(payment, 2)}.')

monthly_mortgage_payment()
```

```python
"""
Three integers—A, B and C—are called a Pythagorean when C * C = A * A + B * B.
Write a program to take an input of integer from the
user and find all the Pythagorean triples of integers, none of which
are bigger than the integer taken from the user. For example, if the
number from the user is 6, then 5, 4, and 3 are a Pythagorean triple
because 5^2 = 4^2 + 3^2 = 25.
"""

def find_pythagorean():
    list_of_triples = []

    # Prompt user for input
    user_number = int(input('Enter your number for testing: '))

    # Iterate through possible values
    for c in range(1, user_number + 1):
        for a in range(1, c):
            for b in range(a + 1, c):
                if a ** 2 + b ** 2 == c ** 2:
                    list_of_triples.append([a, b, c])

    # Print the list of Pythagorean triples
    print(list_of_triples)

# Execute the function
find_pythagorean()
```

```python
"""
Compound interest is a common practice in finance and banking
to earn interest on interest as well as on the principal. Assume that
your bank offers you a deposit account through which you can earn
compound interest, the amount you deposit into the account is p, and
annual compound interest is r. By the end of n years after your initial
deposit, the account balance will be a = p(1 + r)^n. For this project, get
p, r, and n from the user, then calculate and display a table showing
the balance, the interest earned each year, and the total interest
earned so far.
"""

import pandas as pd

def interest_table():
    while True:
        try:
            principal = float(input('Enter your principal without commas or dollar sign: '))
            if principal < 0:
                print('Your principal cannot be less than 0.')
                continue

            rate = float(input('Enter the interest rate (e.g. 0.05): '))
            if rate < 0:
                print('Your interest rate cannot be less than 0.')
                continue

            years = int(input('Enter the number of years on your investment: '))
            if years < 1:
                print("You should have at least one year of investment.")
                continue

            break

        except ValueError:
            print('Invalid entry. Try again.')

    year_list = []
    principal_list = []
    interest_list = []
    cum_interest_list = []

    cum_interest = 0

    for year in range(1, years + 1):
        year_list.append(year)
        principal_list.append(principal)
        
        interest = principal * rate
        interest_list.append(interest)

        principal += interest
        cum_interest += interest
        cum_interest_list.append(cum_interest)

    data = {
        'Year': year_list,
        'Principal': principal_list,
        'Interest': interest_list,
        'Total Interest': cum_interest_list
    }
    table = pd.DataFrame(data)
    print(table)

# Execute the function
interest_table()
```

```python
"""
An integer is called a perfect number* if the sum of its factors,
excluding itself, is equal to the integer itself. For example, 6 is a
perfect number because 6 = 1 + 2 + 3, and 1, 2, and 3 are all its factors.
Write a program to get a number from a user, then determine if the
number is a perfect number. If yes, display all its factors.
"""

def perfect_numbers():
    while True:
        try:
            number = int(input('Enter your number here: '))
            if number < 0:
                print('A perfect number cannot be negative.')
                continue
            break
        except ValueError:
            print('Invalid entry. Try again.')

    factor_list = []

    # Find factors of the number
    for n in range(1, number):
        if number % n == 0:
            factor_list.append(n)

    # Check if the sum of factors equals the number
    if sum(factor_list) == number:
        print(f'{number} is a perfect number. The list of the factors is: {factor_list}.')
    else:
        print(f'{number} is not a perfect number.')

# Execute the function
perfect_numbers()
```

```python
"""
For a given integer n, if another integer m can divide n, then m is
called a factor of n. In mathematics, finding all factors of a given
integer is an important operation, especially for cryptography. Write
a program that takes an integer from the user and determine and
display all the factors of the number given by the user.
"""

def factor_list():
    while True:
        try:
            number = int(input('Enter your integer here: '))
            break
        except ValueError:
            print('Invalid input. Please enter an integer.')

    factors = []

    # Find factors of the number
    for candidate in range(1, number + 1):
        if number % candidate == 0:
            factors.append(candidate)

    # Print the list of factors
    print(f'The factors for {number} are: {factors}')

# Execute the function
factor_list()
```

```python
"""
Read a series of float numbers from the user and calculate and display
the average of all the numbers given by the user. Assume that the
number of float numbers is unknown but that a negative number is
used to indicate the end of the input.
"""

def average_of_floats():
    list_of_floats = []

    while True:
        try:
            number = float(input('Enter your number (enter a negative number to stop): '))
            if number < 0:
                break
            list_of_floats.append(number)
        except ValueError:
            print('Invalid input. Please enter a valid float number.')

    if list_of_floats:  # Check if the list is not empty to avoid division by zero
        avg = sum(list_of_floats) / len(list_of_floats)
        print(f'The average of your numbers: {list_of_floats} is {round(avg, 2)}.')
    else:
        print('No numbers were entered.')

# Execute the function
average_of_floats()
```

```python
"""
For a mortgage with a principal of P at a fixed monthly interest rate of
r that needs to be paid off in Y years or 12 * Y = N months, the monthly
payment would be:
Continuing the project you did above for project 8, calculate a table
showing the interest and principal paid each month and the principal
balance at each month’s end.
"""

import pandas as pd

def mortgage_schedule():
    while True:
        try:
            principal = float(input('Enter your mortgage principal without dollar sign and commas: '))
            if principal < 0:
                print('You cannot have a negative principal.')
                continue

            rate = float(input('Enter your monthly interest rate (e.g., 0.05 for 5%): '))
            if rate < 0:
                print('You cannot have a negative interest rate.')
                continue

            amortization_years = int(input('Enter your amortization period in years: '))
            if amortization_years <= 0:
                print('Amortization period must be greater than zero.')
                continue

            break
        except ValueError:
            print('Invalid entry. Please enter valid numbers.')

    amortization_months = amortization_years * 12

    # Calculate the monthly payment using the formula for fixed-rate mortgages
    numerator = rate * principal * (1 + rate) ** amortization_months
    denominator = (1 + rate) ** amortization_months - 1
    monthly_payment = numerator / denominator

    # Lists to hold schedule details
    starting_balance_list = []
    monthly_payment_list = []
    principal_portion_list = []
    interest_portion_list = []
    ending_balance_list = []

    starting_balance = principal

    for month in range(1, amortization_months + 1):
        interest_portion = starting_balance * rate
        principal_portion = monthly_payment - interest_portion
        ending_balance = starting_balance - principal_portion

        # Append values to respective lists
        starting_balance_list.append(starting_balance)
        monthly_payment_list.append(monthly_payment)
        interest_portion_list.append(interest_portion)
        principal_portion_list.append(principal_portion)
        ending_balance_list.append(ending_balance)

        # Update starting balance for the next month
        starting_balance = ending_balance

    # Create DataFrame for displaying the schedule
    data = {
        'Month': list(range(1, amortization_months + 1)),
        'Starting Balance': starting_balance_list,
        'Monthly Payment': monthly_payment_list,
        'Principal Portion': principal_portion_list,
        'Interest Portion': interest_portion_list,
        'Ending Balance': ending_balance_list
    }
    table = pd.DataFrame(data)
    print(table)

# Execute the function
mortgage_schedule()
```
