```python
"""
1. Write a program that reads a text from a user, then counts and displays
how many words and how many alphanumeric letters are in the text.
"""

def text_counter():
    while True:
        try:
            text = input('Enter your text: ')
            break
        except ValueError:
            print('Invalid Input.')

    word_list = text.split(' ')
    alphanumeric_letter_count = 0

    for char in text:
        if char.isalnum():
            alphanumeric_letter_count += 1

    print(f'There are {len(word_list)} words, and {alphanumeric_letter_count} alpha-numeric characters in the text.')

text_counter()
```

```python
"""
2. Write a program that
a. reads a series of numbers that are separated by whitespace and
uses a new line to end the input, then converts the numbers in the
input string and puts them into a list.
b. sorts the numbers in the list in descending order, using the sort()
list object method.
c. sorts the numbers in the list in descending order, using the Python
built-in
function sorted().
Write your own code to sort the numbers in the list in ascending order
without using the sort() method or sorted() function.
"""

def sort_numbers():
    while True:
        try:
            user_numbers = input('Enter the series of integer numbers using only a whitespace to separate each number.')
            if ',' in user_numbers or '.' in user_numbers: # if ',' or '.' in user_numbers won't work because ',' is a non-empty string and will always result in True.
                print('Use only whitespaces to separate the numbers.')
            else:
                break
        except ValueError:
            print('Invalid Input.')
    
    user_numbers_list = user_numbers.split(' ')
    user_numbers_int = [int(num) for num in user_numbers_list if num]

    def manual_sort(lst):
        sorted_lst = [lst[0]]
        for element in lst[1:]:
            flag = True
            for i in range(len(sorted_lst)):
                if element <= sorted_lst[i]:
                    sorted_lst.insert(i, element)
                    flag = False
                    break
                if flag == True:
                    sorted_lst.append(element)
        return sorted_lst
    
    sort_with_manual_function = manual_sort(user_numbers_int)
    sort_with_sorted = sorted(user_numbers_int, reverse=False) # Creates a new list
    user_numbers_int.sort(reverse=False) # Modifies the existing list

    print(sort_with_manual_function, sort_with_sorted, user_numbers_int)

sort_numbers()
```

```python
"""
4. Every course offered at universities has a course number and a title.
For this project, write an application that uses a dictionary to save
course information, allows users to add a course into the dictionary,
and allows a user to get the title of a course for a given course
number. The application should perform the following functions:
a. Get a course number and name from a user and add an item, with
the course number as key and the name as value, to the dictionary
if the course doesn’t already exist in the dictionary.
b. Get a course number from a user, then find out the name of the
course.
c. Display a list of all the courses in the dictionary showing the course
numbers as well as names.
d. Quit the application.
Hint: You will need a top-level
while loop, which displays a menu
showing the four options then acts accordingly.
"""

# Add a course
# Get a course number to find out the name of the course
# Display a list of all the courses
# Quit the application

def course_application():
    print(f'Welcome to the course app.')
    course_details = {}

    while True:
        try:
            selection = input(f'Please choose from the below options:\na) Add a course\nb) Find the name of the course\nc) Display all courses\nd) Quit').lower()
            if selection not in ['a','b','c','d']:
                print('Invalid selection. Choose an option from a, b, c, d.')
            else:
                if selection == 'a':
                    while True:
                        try:
                            k = input('Enter the course number: ')
                            if not k or k in course_details.keys():
                                print('You cannot have an empty course number or the course number cannot already exist.')
                            else:
                                break
                        except ValueError:
                            print('Invalid input.')
                    while True:
                        try:
                            v = input('Enter the course name: ')
                            if not v:
                                print('You cannot have an empty course name.')
                            else:
                                break
                        except ValueError:
                            print('Invalid Input')
                    course_details[k] = v

                elif selection == 'b':
                    while True:
                        try:
                            course_number = input('Enter the course number you would like to explore: ')
                            if course_number in course_details.keys():
                                print(f'The course name for {course_number} is {course_details[course_number]}.')
                                break
                            else:
                                print('The course number does not exist.')
                        except ValueError:
                            print('Invalid input.')
                
                elif selection == 'c':
                    for key, value in course_details.items():
                        print(f'{key}: {value}')
                    
                else:
                    print('Thank you for using. Bye.')
                    break                
        except ValueError:
            print('Invalid selection. Choose an option from a, b, c, d.')

course_application()
```

```python
""" 
6. Cryptography is the study of theory and technology for the protection
of confidential documents in transmission or storage. It involves both
encryption and decryption. In any cryptographic scheme, encryption
is the process of converting plaintext to ciphertext according to a
given algorithm using an encryption key, whereas decryption is the
process of converting encrypted text (ciphertext) back to plaintext
according to a given algorithm using a decryption key. If the
encryption key and decryption key are the same in a cryptographic
scheme, the scheme is a symmetric cryptographic scheme; if the
two keys are different, the scheme is an asymmetrical cryptographic
scheme.
Among the many cryptographic schemes, substitution is a classic
one, though the scheme is prone to frequency analysis attack. Write a
program that can
a. automatically generate a substitution key and add it to a key list
stored in a file.
b. display the substitution keys in the file.
c. allow the user to choose a key in the list and encrypt some text
taken from the user.
d. allow the user to choose a key to decrypt an encrypted text taken
from the user.
e. allow the user to choose a key, encrypt the content of a text file,
and save the encrypted content into a different file.
f. allow a user to choose a key, decrypt the encrypted content in a
file, and display the plaintext decrypted content.
"""
import random
random.seed(1)
import string
import os
# random.shuffle() directly modifies the list you pass to it, instead of returning a new list. 
# After calling this function, the original list will be in a randomly shuffled order.
# string.ascii_uppercase returns one string consisting of all uppercase alphabets

def initiate_cryptography():
    upper_case = list(string.ascii_uppercase)
    lower_case = list(string.ascii_lowercase)

    all_case = upper_case + lower_case

    new_case = all_case.copy()
    random.shuffle(new_case)

    all_case_str = ''.join(all_case)
    new_case_str = ''.join(new_case)

    file_path = os.path.join('Texts', 'cryptography.txt')
    with open(file_path, 'w') as w:
        w.write(all_case_str)
        w.write('\n')
        w.write(new_case_str)



def encrypt():
    file_path = os.path.join('Texts', 'cryptography.txt')
    with open(file_path, 'r') as r:
        content = r.read()

        content_list = content.split('\n')

        norm_alpha = list(content_list[0])
        encrypt_alpha = list(content_list[1])

        crypto_dict = dict(zip(norm_alpha, encrypt_alpha)) # Zip the two lists so that each element is mapped to the corresponding new element. Also, remember that dictionary keys must be immutable; lists are mutable.
        
    str_to_encrypt = input('Enter a text to encrypt: ')
    
    encrypted_code = ''

    for char in str_to_encrypt:
        if char in crypto_dict.keys():
            encrypted_code += crypto_dict[char]
        else:
            encrypted_code += char
    
    print(encrypted_code)


encrypt()

```
